# Reprezentarea grafică a diferitelor tipuri de programe liniare

- Incompatibile (fără soluție)
- Compatibile:
  - Cu optim infinit
  - Cu optim unic, finită
  - Cu o infinitate de soluții optime

Voi folosi `numpy` și `matplotlib` pentru a reprezenta grafic sistemele.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Sistem incompatibil

Luăm ca exemplu următorul program liniar:
$$
\max x + y \\
x + 2y \geq 3 \\
x - 2y \geq 2 \\
y \leq -2
$$

Pentru a face afișarea, scriu expresiile de mai sus cu $y$ în funcție de $x$:
$$
y = x \\
y = \frac{x - 3}{2} \\
y = - \frac{x - 2}{2} \\
y = -2
$$

```python
x = np.linspace(-5, 5, 200)

plt.figure(dpi=100)

plt.suptitle("Program liniar incompatibil")

plt.xlim(-3, 5)
plt.ylim(-4, 4)

# Funcția obiectiv, `x + y`
plt.plot(x, x, color='red')

# x + 2y >= 3
plt.plot(x, (x - 3)/2)
plt.fill_between(x, (x - 3)/2, (x - 3)/2 + 10, alpha=0.2)

# x - 2y >= 2
plt.plot(x, -(x - 2)/2)
plt.fill_between(x, -(x - 2)/2, -(x - 2)/2 + 10, alpha=0.2)

# y <= -2
plt.plot(x, x * 0 - 2)
plt.fill_between(x, -2, -2 - 10, alpha=0.2)

plt.show()
```

După cum se vede și în grafic, nu există o zonă în care toate cele 3 semiplane să se intersecteze simultan. Deci nu avem unde să căutăm valoarea maximă a lui `x + y`.
