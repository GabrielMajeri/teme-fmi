# Reprezentarea grafică a diferitelor tipuri de programe liniare

Un program liniar poate fi:

- Incompatibil (fără soluție)
- Compatibil:
  - Cu o soluție optimă (de valoare finită)
  - Cu o infinitate de soluții optime (de aceeași valoare)
  - Cu optim infinit (nemărginită)

Voi folosi `numpy` și `matplotlib` pentru a reprezenta grafic sistemele.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Sistem incompatibil

Luăm ca exemplu următorul program liniar:
$$
\max x + y \\
x + 2y \geq 3 \\
x - 2y \geq 2 \\
y \leq -2
$$

Pentru a face afișarea, scriu expresiile de mai sus cu $y$ în funcție de $x$:
$$
y = \frac{3 - x}{2} \\
y = \frac{x - 2}{2} \\
y = -2
$$

```python
x = np.linspace(-5, 5, 200)

plt.figure(dpi=100)

plt.xlim(-3, 5)
plt.ylim(-4, 4)

# x + 2y >= 3
y = (3 - x)/2
plt.plot(x, y)
plt.fill_between(x, y, y + 10, alpha=0.2)

# x - 2y >= 2
y = (x - 2)/2
plt.plot(x, y)
plt.fill_between(x, y, y + 10, alpha=0.2)

# y <= -2
y = x * 0 - 2
plt.plot(x, y)
plt.fill_between(x, y, y - 10, alpha=0.2)

plt.show()
```

După cum se vede și în grafic, nu există o zonă în care toate cele 3 semiplane să se intersecteze simultan. Deci nu avem unde să căutăm valoarea maximă a lui $x + y$.

## Sistem cu soluție optimă finită

$$
\max x + y \\
x - 2y \geq -5 \\
x + y \leq 3
$$

```python
x = np.linspace(-5, 5, 200)

plt.figure(dpi=100)

plt.xlim(-3, 5)
plt.ylim(-4, 4)

# x - 2y >= -5
y = (x + 5)/2
plt.plot(x, y)
plt.fill_between(x, y, y -10, alpha=0.2)

# x + y <= 3
y = 3 - x
plt.plot(x, y)
plt.fill_between(x, y, y - 10, alpha=0.2)

# Punct de optim
plt.scatter(1/3, 8/3, c='red', zorder=5)

plt.show()
```

Cea mai mare valoare a lui $x + y$ se poate obține în punctul de intersecție ale dreptelor $x - 2y = 5$ și $x + y = 3$.

## Sistem cu optim unic, infinit

$$
\max x + y \\
x + y <= 5
$$

```python
x = np.linspace(-5, 5, 200)

plt.figure(dpi=100)

plt.xlim(-5, 5)
plt.ylim(-5, 5)


# x + y <= 5
y = 5 - x
plt.plot(x, y)
plt.fill_between(x, y, y - 15, alpha=0.2)

plt.show()
```

Observăm că toate punctele de pe dreapta $x + y = 5$ maximizează funcția obiectiv $x + y$, deci avem o infinitate de soluții optime.

## Sistem cu optim infinit

$$
\max x + y \\
x \geq 0
$$

```python
plt.figure(dpi=100)

plt.xlim(-5, 5)
plt.ylim(-5, 5)

# x >= 0
y = np.linspace(-5, 5, 200)
plt.plot(0 * y, y)
plt.fill_betweenx(y, 0, 5, alpha=0.2)

plt.show()
```

Putem să luăm orice pereche de puncte $(x, y)$ cu $x \geq 0$, și putem obține o valoare oricât de mare vrem pentru funcția obiectiv $x + y$.
